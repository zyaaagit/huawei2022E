```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#数据导入
dateparse=lambda dates: pd.datetime.strptime(dates,'%Y')
data3 = pd.read_excel("放牧强度与化学性质.xlsx",index_col='year',date_parser=dateparse)
data3 = data3.iloc[:,:7]
```

```python
data3_NG = data3[data3["放牧强度（intensity）"] =="NG" ]
data3_LGI = data3[data3["放牧强度（intensity）"] =="LGI" ]
data3_MGI = data3[data3["放牧强度（intensity）"] =="MGI" ]
data3_HGI = data3[data3["放牧强度（intensity）"] =="HGI" ]
```

```python
data3_NG_G17 = data3_NG[data3_NG['放牧小区（plot）'] == 'G17']
data3_NG_G19 = data3_NG[data3_NG['放牧小区（plot）'] == 'G19']
data3_NG_G21 = data3_NG[data3_NG['放牧小区（plot）'] == 'G21']
plt.rcParams['font.sans-serif'] = ['SimHei']  # 黑体
plt.rcParams['axes.unicode_minus'] = False    # 解决无法显示符号的问题
sns.set(font='SimHei', font_scale=1)        # 解决Seaborn中文显示问题
plt.subplots(figsize=(20,4),dpi=140,facecolor='w')# 设置画布大小，分辨率，和底色

plt.subplot(131)
plt.plot(data3_NG_G17.groupby('year').mean())
plt.legend(labels=['SOC','SIC','STC', 'N','C/N'],loc=1)
plt.xlabel('G17')
plt.ylabel('NG:      ', rotation=0,fontdict={'size': 16})

plt.subplot(132)
plt.plot(data3_NG_G19.groupby('year').mean())
plt.legend(labels=['SOC','SIC','STC', 'N','C/N'],loc=1)
plt.xlabel('G19')

plt.subplot(133)
plt.plot(data3_NG_G21.groupby('year').mean())
plt.legend(labels=['SOC','SIC','STC', 'N','C/N'],loc=1)
plt.xlabel('G21')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel("test3.xlsx")
```

```python
data.head()
x = data.columns[1:5]
y_NG = data.iloc[0,1:]
y_LGI = data.iloc[1,1:]
y_MGI = data.iloc[2,1:]
y_HGI = data.iloc[3,1:]
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei']  # é»‘ä½“
plt.rcParams['axes.unicode_minus'] = False    # è§£å†³æ— æ³•æ˜¾ç¤ºç¬¦å·çš„é—®é¢˜
sns.set(font='SimHei', font_scale=1)        # è§£å†³Seabornä¸­æ–‡æ˜¾ç¤ºé—®é¢˜
plt.subplots(figsize=(6,4),dpi=1080,facecolor='w')# è®¾ç½®ç”»å¸ƒå¤§å°ï¼Œåˆ†è¾¨çŽ‡ï¼Œå’Œåº•è‰²

x_width = range(0,len(x))
x2_width = [i+0.15 for i in x_width]
x3_width = [i+0.3 for i in x_width]
x4_width = [i+0.45 for i in x_width]

plt.bar(x_width,y_NG,width = 0.15, color="w",edgecolor="k",hatch="///",label = 'NG')
plt.bar(x2_width,y_LGI,width = 0.15,color="w",edgecolor="k",hatch="xxx",label = 'LGI')
plt.bar(x3_width,y_MGI,width = 0.15,color="gray",edgecolor="k",label = 'MGI')
plt.bar(x4_width,y_HGI,width = 0.15,color="black",edgecolor="k",label = 'HGI')
plt.legend(loc=3)
plt.xticks(range(0,4),x)
```
